```python
def karatsuba(num1,num2):
    n1 = str(num1)
    n2 = str(num2)
    
    a = int(n1[0:len(n1)//2])
    b = int(n1[len(n1)//2:])
    c = int(n2[0:len(n2)//2])
    d = int(n2[len(n2)//2:])
    
    a0 = a*c
    a1 = (a-b)*(c-d)
    a2 = b*d
    
    part1 = a0*10**int(len(str(num1)))
    part2 = (a0+a2-a1)*10**int(len(str(num1))//2)
    part3 = a2
    
    r = part1 + part2 + part3
    
    return r
```
